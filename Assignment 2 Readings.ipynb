# Definition of capital in Basel III - Bank For International Settlements

### Executive Summary
* Basel III reforms December of 2010. 
* Basically increase more capital and stringent disclosure requirements. 
* Focuses on high quality capital => shares and retained earnings that can absorb losses. 
* Going- and gone- concern framework (Tier 1 Capital and Tier 2 Capital Respectively). More explained later.
* These capital instruments must be able to absorb losses at a point of non-viability (PoNV) before taxpayers are exposed to loss. 
* Enhanced disclosure requirements meant to improve the transparency of banks' capital bases and improving market discipline. Reduce inconsistencies in its implementation across jurisidictions. 

### Components of Regulatory Capital

__*TIER 1 CAPITAL*__

---
Going-concern Capital

__Common Equity Tier 1 Capital (CET1)__ 

Highest quality of regulatory capital and absorbs losses when they occur. 

__Additional Tier 1 Capital (AT1)__

Provides loss absorption on a going-concern basis. 
Does not meet same criteria as CET1.

EG: Debt instruments such as perpetual contingent convertible capital instruments are included in AT1 but not CET1. 

* Note: Contingent Convertible Capital Instruments (CoCos):  CoCos are hybrid capital securities that absorb losses in accordance with their contractual terms when the capital of the issuing bank falls below a certain level. 2 Then debt is reduced and bank capitalisation gets a boost.

__*TIER 2 CAPITAL*__

---
Gone-concern Capital

When a bank fails, or after it has 'gone', Tier 2 instruments absorb lossess before depositors and general creditors do. 

* Criteria for inclusion less strict than AT1. 
* Allows for instruments with a maturity date to be Tier 2 while perpetual instruments are eligible for AT1. 

__*TIER 1 + TIER 2 = Regulatory Capital*__

* Banks are required to maintaing specified amount of CET1 and total capital as a % of risk-weighted assets (NOT TOTAL ASSETS).

\* The standard requires instruments issued by non-joint stock companies to meet a set of criteria to be deemed equivalent to common
shares and included in CET1.

* PoNV requires that AT1 and Tier 2 to be capable of being converted to common equity or written off. 

* Triggers for writeoffs/conversion: 

    * (i) a decision of the relevant authority that the conversion/write-off is necessary, given that the bank is assessed to be non-viable;
    
    * (ii) a decision to inject public funds to prevent the bank’s failure. 
    
This may happen based on either theauthority’s statutory powers or the contractual features of the capital instruments. 

__Minority (or non-controlling) interest__

*Essentially, minority interest in other companies can be part of any of the three types of capital. 

* Minority interest, or non-controlling interest, arises from capital instruments issued to third parties by a fully consolidated subsidiary of a bank and can be located in any of the three components of regulatory capital: CET1, AT1 and Tier 2

* If issued by the parent bank, it would meet all criteria for regulatory capital. 

* In addition, the minority interest must not be funded, either directly or indirectly, by the parent bank and must be issued by a subsidiary that is itself a bank.

The key note in this section: 

* The amount of capital must be able to support the consolidated group. Any surplus capital caan be repaid to non-controlling interest.
* However, it limits the recognition of min interest to cover capital requirements and excludes surplus capital of the subsidiary that is attributable to non-controlling interest. Basically, surplus capital of the subsidiary cannot be used to cover capital requirements of the parent. 

__Regulatory Adjustments__

* Regulatory adjustments and deductions from regulatory capital are based on the high degree of uncertainty of positive realisable value in periods of stress, mostly related to CET1 capital.

    * Includes: Goodwill, intangible assets, deferred tax assets, investments, and others. 
    
    
    
* Investments of banks into other financial institutions should not be "double counted"...aka minority interest to cover regulatory capital should not be hinged on the value/expected realisable value of the interest in another bank/financial inst. Therefore, it should be deducted. Referred to as Corresponding Deduction Approach. To be more specific: 

    * Deduction to be applied by the investing bank. 
    
    
* New thresholds are applied to certain of the deductions under Basel III. Banks receive limited recognition for their significant investments in the common shares of unconsolidated financial institutions, mortgage servicing rights and deferred tax assets that arise from temporary differences. Each item is individually capped at 10% of the bank’s common equity after the application of certain regulatory adjustments, and the aggregate is limited to 15% of the bank’s CET1. Non-significant investments in unconsolidated financial institutions (that is, where the bank owns less than 10% of the common shares) are only deducted to the extent that all such exposures in aggregate exceed 10% of the bank’s common equity. 

   * Basically means that investments in a subsidiary of a financial institution has limited recognition as regulatory capital. Capped at 10% of bank's common equity after adjustments and aggregate is limited to 15% of CET1. If bank owns less than 10% of common shares in an unconsolidated financial institution, they are deducted up to the extent that exposures exceed 10% of bank's common equity.
