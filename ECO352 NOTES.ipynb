Aggregate liquidity can be understood as the rate of growth of the aggregate financial sector balance sheet. When asset prices increase, financial intermediaries’ balance sheets generally become

their surplus capital. In analogy with manufacturing firms, we may see the financial system as having

## Bank activity and funding strategies: The impact on risk and returns
* Authors: Asli Demirguc-Kunt, Harry Huizinga 
* Publication: Jounral of Financial Economics

### Abstract

This paper examines the implications of bank activity and short-term funding strategies for bank risk and return using an international sample of 1,334 banks in 101 countries leading up to the 2008 financial crisis. Expansion into noninterest income-generating activities such as trading increases the rate of return on assets, and it could offer some risk diversification benefits at very low levels. Nondeposit, wholesale funding in contrast lowers the rate of return on assets, while it can offer some risk reduction at commonly observed low levels of nondeposit funding. A sizable proportion of banks, however, attract most of their short-term funding in the form of nondeposits at a cost of enhanced bank fragility. Overall, banking strategies that rely prominently on generating noninterest income or attracting nondeposit funding are very risky, consistent with the demise of the US investment banking sector.

* Funding short-term strategies with trading increases inherent risk of the assets being funded. 
* This comes at the cost of bank fragility. 

### Summary

* Banks have diversified their product offerings by consolidating businesses that were otherwise independent. For example, investment banks stood alone in their own business. Now, we see investment banks owned by large banks, along with commercial banks and insurance firms. This allows to diversify against adverse shocks. 

* Banks that were the most deeply in trouble were the ones that generated substantial noninterest income through trading of, for instance, MBS or relied on wholesale capital markets for their funding.

* Large fast growing banks tend to have higher non-interest income and nondeposit funding shares. 

* Paper objective is to provide evidence on what funding strategies and bank income perform well in terms of producing profitable and stable banks. 

* Suggests that low levels of noninterest income and nondeposit funding, there oculd be some risk diversification benefits of icnreases these shares. Higher levels of noninterest income and nondeposit shares, further increases bank risk. 

* IV Regression: Paper instruments for bank income and funding mixes as a proxy for bank-specific banking powers and allowed activities. Risk outcomes and funding strategies and return outcomes can be determined simultaneously. 
    * Results show that higher noninterest income or nondeposit funding share engenders a materially higher bank risk, but the impact of either variable on the ROA is difficult to establish due to endogeneity. 
    
* Banking strategies that rely on noninterest income or nondeposit funding are very risky.

## CoCos: a primer
* Authors: Stefan Avdjiev, Anastasia Kartasheva, Bilyana Bogdanova
* Publication: Bank for International Settlements

### Abstract / Executive Summary

Contingent convertible capital instruments (CoCos) are hybrid capital securities that absorb losses when the capital of the issuing bank falls below a certain level. In this article, we go over the structure of CoCos, trace the evolution of their issuance, and examine their pricing in primary and secondary markets. CoCo issuance is primarily driven by their potential to satisfy regulatory capital requirements. The bulk of the demand for CoCos has come from small
investors, while institutional investors have been relatively restrained so far. The spreads of CoCos over other subordinated debt greatly depend on their two main design characteristics – the trigger level and the loss absorption mechanism. CoCo spreads are more correlated with the spreads of other subordinated debt than with CDS spreads and equity prices. 

### Introduction

* Cocos address the problem of a lack of incentive for private investors to provide additional capital in times of distress, which at that point governments use tax-payer money and distorts the incentives of bankers. 
* When the capital of the issuing bank falls below a certain level, debt level is lowered. 
* Looks at dataset that covers 70bn worth of CoCos issued between June 2009 and June 2013. 

### Reasons for issuing CoCos 

1. Satisfies regulatory capital requirements 
2. Demand comes from private banks and retail investors. 
3. Tend to yield more thna higher ranked debt and are highly dependent on trigger level and loss absorption. 
4. CoCos tend to be more correlated with those of other subordinated debt than with CDS spreads and equity prices. 

### Structure and design of CoCos 

__Requirements__ 
* They need to absorb lossess prior or at the point of insolvency. 
* The loss absorption mechanism must be a function fo the capitalisation levels of thee issuing bank. 
* Robust to price manipulation and speculative attacks. 

__TWO MAIN DEFINING CHARACTERISTICS__ 
1. Loss absorption mechanism
    * Converting into common equity or by suffering a principal writedown. 
2. Trigger to activate that mechanism 
    * Trigger can be mechanical (defined in terms of capital ratio) or discretionary (supervisory judgement).
    
### Triggers

* Based on mechanical rules or supervisors' discretion.
    * EG: Capital falls below a certain % of RWA. Can be based on book value or market value. 

_Accounting Triggers_: Book value triggers. Set contractually based on BV of CET1 capital as a ratio of RWA. Depends heavily on the frequency of disclosure and calculations. May not be activated in a timely manner. 

_Market-Value Trigger_: Set to min ratio at the bank's stock market cap to its assets. Reduces scope for sheet manipulation and regulatory forbearance. Subject to stock price manipulation. Dilutive conversion rate makes it possible for more than one pair of equity prices to exist for any combination of bank asset values and non-CoCo debt levels => double equilibria problem. There is also an incentive to shortsell underlying asset to trigger. 

_Discretionary Triggers_: Point of non-viability (PONV) triggers are triggered based on supervisors' judgement about the bank's solvency prospects. Unless disclosure is made clear, uncertainty around timing looms over.

### Loss absorption Mechanism

CoCos boost the bank's equity in two ways: 

1. Conversion-to-equity CoCo: Turns into CET1 at a pre-defined conversion rate. 
    i. Conversion rate based on market price of stock at time triggered.
    ii. Pre-specified price. 
    iii. Combination of the two. 

2. Principal Writedown (PWD): CoCo raises equity by incurring a writedown. 
    * Full or partial writedown feature. EG: 75% FV and 25% cash. 

We note that CE CoCos with a combination would incentivize existing equity holders to avoid a breach of the trigger while preventing unlimited dilution. (i causes dilution, ii lacks incentive to avoid triggers).

### CoCo Issuance

* Under Basel III, qualifies as either AT1 or T2 capital. 
    * Requires a PONV trigger requirement. 
    * Requires a going-concern contingent capital requirement, which only applies to AT1 instruments classified as liabilities. 

![Basel III CoCos](CoCo_BaselIII.png "CoCo Basel III")

* Selection for trigger is a tradeoff between regulatory capital eligibility considerations and costs of issuance. 

* Lower triggers => usually less likely to qualify as AT1 capital (since lower loss absorbency). 

__Minimum trigger as AT1 capital is 5.125%__

* AT1 instruments must be perpetual, that is why a third of all CoCos have no maturity date. Dated maturity CoCos are T2. 

__Capital eligibility considerations not as important in the selection of loss absorption mechanisms__ 

* CE is more popular.
    * Usually cheaper issuance than PWD. 
* Tax implications across jurisdictions vary with CoCos. Some see it as debt, in which interest expense is associated. 

* 64% of CoCos are tax-deductible, 20% are not, 16% under review. 

### Investors in CoCos

### Primary market pricing of CoCos

Main Determinants of Pricing: 

1. Trigger
2. Loss absorption mechanism
3. Capital structure of the bank 

* Yields are consistent with bank capital structure.

__High triggeer tends to yield more than low-trigger cocos.__

## Contingent Capital and Bail-In Debt: Tools for Bank Resolution. 

* Authors: Chris D'Souza and Toni Gravelle from BOC, Walter Engert and Liane Orsi from OSFI. 
* Publication: Bank of Canada and OSFI

### Introduction

* Policy makers were unwilling to allow institutions to fail. 
* Concerns arose regarding excessive risk-taking behaviour by protected institutions. 
* Paper looks at two potential resolutions of failing banks: contingent capital and bail-in debt. 

### What are Contingent Capital and Bail-In Debt? 

_Contingent Capital_: Subordinated security, similar to that of a preferred share or subordinated debenture, that converts to common equity under certain conditions. => Conversion terms are set out in the contracts made between the issuer and investor. Require payment for taking on the risk. 

_Bail-In Debt_: Contractual mechanism extended beyond the regulatory capital base to senior debt securities of the issuing bank.

__Gone-Concern and Going-concern Contingent Instruments__

Contingent Capital is categorized into two main types of instruments: 

_Gone-Concern_: Supervisor determines whether the financial condition (or PONV) warrents conversion to common equity. Offers itself to the resolution framework. 

_Going-Concern_: Converts before PoNV. Usually if capital ratio hits a certain point or share price falls below a certain level or aggregate banking sector falls. 


### Regulatory context matters

With gone-concern capital, the incentives of the supervisor and resolution authority to intervene in a timely manner matters. 

Going-concern => works to exercise forebearance. Tendency of som esupervisors to accomodate troubled institutions and to refrain from intervening with corrective measures. Usually happens due to lack of incentives or power to intervene effectively as the bank deteriorates. Basically, reduces supervisory forebearance. 

### Objectives of gone-concern contingent instruments 

Gone-concern contingent capital and bain-in debt share two objectives: 

(i) To support the resolution of a failing bank by providing sources of capital when the institution cannot recapitalize through private markets; and 

(ii) to ensure that equity holders and other providers of regulatory capital, as well as major creditors of banks, face risk of loss, even if the troubled bank is not closed and liquidated. 

These mechanisms can reduce the liklihood of a bailout and reduce cost to taxpayers. 

### How would gone-concern contingent instruments help resolution? 

We first note the conditions for conversion: 

(i) Reach PoNV; or

(ii) Public sector provides capital, without which the bank would not be viable. 

Gone-concern capital might not be enough to generate sufficient capital. 

Looking at bail-in debt: 
* Could increase the private sector capital available at PoNV. 
* Most senior debt would tend to be unchanged if conversion triggered. 
* We could use both bail-in and contingent instruments, alongside existing prudential capital requirements, for restoration requirements. 

### Non-viability and supervisory discretion

The OSFI can intervene and essentially act as a supervisor if, under discretion, finds a bank to be failing. Can activate trigger on their own discretion. 

### Conclusion

Several issues concerning contingent capital and bail-in senior debt require further analysis, including the precise form of any prudential rules to bring these mechanisms into effect; the amount of such contingent instruments to require; and the determination of the period over which such requirements would become binding. Other issues include how these mechanisms would relate to other tools in the regulators’ resolution toolkit; their incentive effects on market participants, including when conversion seems likely; and the implications for investors holding instruments that are not subject to conversion.

These issues should be addressed as policy-makers around the world continue to debate the set of tools to be included in an effective framework for bank resolution. The framework should recognize the possibility that financial firms can fail, and that they need to be resolved efficiently, in a way that minimizes disruption to the wider economy and reduces the risk of moral hazard. Relying on higher prudential requirements for common equity is undoubtedly part of the solution, as was recently proposed by the Basel Committee. In addition, contingent capital and bail-in debt could improve the capacity of the private sector to contribute to the resolution of failing banks while reducing risks to the public sector and improving the incentives that condition market behaviour.

### What is liquidity at a bank? 

* The ability to find cash it may need to meet demands upon it. 
* Currency or cash or on account at the Fed Res or Central Bank
* Largest commitment is to return demand deposits. 

### Why do we care about it? 

* Banks are fragile in nature without sufficient safety margins (capital ratio, SLR, etc). 
* Asset illiquidity (credit freeze) cannot mitigate the risk/loss associated with fire-sales and bank runs. 

### Why are banks prone to runs? 
* Banks purpose is maturity transformation, also known as time intermediation.

__time intermediation__: Take deposits and other short-term funds and lend them back out at longer-maturities. 

* Generally, banks can meet this obligation. 
* Problem arises with lost confidence, "en masse". 
* CDIC helps mitigate the risk of a bank run by restoring a level of confidence during times of financial distress. 
* Modern banks heavily rely on short-term lending, or time intermediation, and so borrowings/liquidity in capital markets matter. 
* Being unable to roll over short-term loans becomes a problem. 

### How can banks achieve adequate liquidity?

1. Shorten asset maturities
    * Shorter maturities allow for meeting cash demands during periods of cash-crunches. 
2. Improve the average liquidity of assets
    * Credit worthy securities and those with higher volumes tend to be more liquid. Securities tend to be more liquid, with associated risk of course. 
3. Lengthen liability maturities
    * Basically, extend the loans you owe etc. 
4. Issue more equity
    * Common stock = perpeutual maturity. 
    * No interest needs to be paid
5. Reduce contingent commitments
    * Cut back on borrowing volumes of lines of credits etc. 
6. Obtain liquidity protection
    * Can pay another bank or insurer for the guarantee of cash in the future, collateralized to the bank's assets. 

__Yield Curves__

_Upward Sloping Yield Curve_: This is the most typical shape of the yield curve. This suggests that shortening maturities decreases investment income and lengthening liability maturities raises interest expense. More liquid instruments tend to have lower yields => reduce investment income.

### How much liquidity is enough.

### What is the economic value of maturity transformation? 

* Note higher liquidity will result in higher safety, but more fragility of the banks.

* Why do we need maturity transformation? It serves to our preferences for liquidity and need for longer-term investment funds. 

_Diamond and Dybvig(1983)_: Case for maturity transformation - driven by desire of consumers and businesses for a high level of liquidity for a significant part of their assets. Allowing investors to invest in illiquid assets, while allowing them to withdraw at once is a productive activity. Allows the economy to benefit without foregoing illiquid investments.

### How does the liquidity coverage ratio work 

$$ LCR = \frac{\text{high quality liquid assets}}{\text{projected cash claims over the next 30 days}} $$

* Specifies the percentage of liabilities with an indefinite maturity.
