__B. Assessing the Effectiveness of Macroprudential Measures__
* Estimate two separate equations for mortgage credit and house prices.

$$ Y_t = \alpha + \beta X_t + \gamma D_{t}^i + \epsilon_t $$

    * $Y_t$ is mortgage credit or house price YoY growth.
    * $X_t$ is a matrix of control variables (Current and lagged)
    * $D_{t}^i$ is a dummy variable equal to 1 in months following implementation of a set of measure i where i represents a specific set of macroprudential measures. 
* The model found that 2nd,3rd, and 4th rounds of measures helped to limit the increase in household leverage and house prices. 
* Evidence suggests that last 3 rounds of measures dampened mortgage credit growth aand house prices. 
* Reducing mortgage growth, tightening of mortgage insurance rules also dampened house price growth. 
* Tightening LTVs for new mortgage and for refinancing loans had the largest impact.

__What is a repo and why does it matter?__ 

* Repurchase agreement: fin inst sells security on understanding it will buy it back at a pre-agreed price on a fixed future date. 

* Essential for collateralized loan. Interest on loan in excess of repo price over the sale price. 
* From the perspective of the lender, the party who buys the security with the undertaking to re-sell it later is called a reverse repo. 
* From the perspective of the buyer, it is equivalent to granting a loan on collateral. 
* Why does this increase liquidity? As a dealer who takes a short position will borrow the needed security and provide cash or other securities as collateral. Closes out position by buying the borrowed security. In the case of a long, if a broker cannot purchase a position in equity, they will borrow  and usually uses the purchased security as collateral for the cash borrowing. 

__Value at Risk__ 

$$ Prob (A < A_0 - VaR) \leq 1-c$$

* A is a RV
* c is th econfidence interval
* $A_0$ is the smallest non-negative number VaR such that the above holds true. 

For example: A can be the total marked to marked assets, which then the equation tells us the equity capital that the firm must hold to stay solvent with a probability c. 

VaR is informative of leverage. Paper conjectures:

$$ K = \lambda \times VaR$$

* Lambda is the proportion of capital that intermediary holds per unit of VaR. 

Leverage satisfies: 

$$ L = \frac{A}{K} = \frac{1}{\lambda} \times \frac{A}{VaR} = \frac{1}{\lambda}\times \frac{1}{V}$$

* V is the unit value at risk.

So we see that leverage increases as the total assets per unit of VaR decreases. 

* This effect is negative with leverage (if it is not clear already)

__Conclusion__ 

Aggregate liquidity can be understood as the rate of growth of the aggregate financial sector balance sheet. When asset prices increase, financial intermediaries’ balance sheets generally become
stronger, and – without adjusting asset holdings – their leverage tends to be too low. The financial
intermediaries then hold surplus capital, and they will attempt to find ways in which they can employ
their surplus capital. In analogy with manufacturing firms, we may see the financial system as having
‘‘surplus capacity”. For such surplus capacity to be utilized, the intermediaries must expand their balance sheets. On the liability side, they take on more short-term debt. On the asset side, they search for
potential borrowers. Aggregate liquidity is intimately tied to how hard the financial intermediaries
search for borrowers. In the sub-prime mortgage market in the United States we have seen that when
balance sheets are expanding fast enough, even borrowers that do not have the means to repay are
granted credit – so intense is the urge to employ surplus capital. The seeds of the subsequent downturn in the credit cycle are thus sown.
