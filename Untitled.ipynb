# Major Assignment Readings

As of Feb 9th 2015, LTV limits introduced. 

The objectives are to: 

1. Dampen the pro-cyclical dynamics that could occur between property lending and house prices. 

2. Strengthen resilience of banks and households to financial shocks. 

These measures were introduced in a precautionary way to ensure that housing-market recovery would not be destabilised with a re-emergence of housing market speculation driven by credit. 

Many countries implemented LTV DTI or LTI restrictions. 

__What is the policy__ 
* 80% LTV cap for NFTB (non-first-time buyers) of PDHs (primary welling homes). 
* For first-time buyers (FTBs), a higher cap of 90% applies for the first 220,000 of the value of the home and then 80% LTV for anything above 220,000. 
* No more than 15% of new lending for PDHs is allowed above the caps. 
* 70% LTV for BTL (buy-to-let) purchases. 
* LTI maximum is 3.5. Up to 20% of new lending is allowed above this limit. 
* No LTI limit imposed on BTL and only applies to primary residences. 

__Exceptions to the measures__ 
1. Borrowers with negative equity and selling their home and purchasing a new one are exempt from LTV limits. 

2. Switcher mortgages (borrowers changing mortgage provider with no increase in principal. 

3. Mortgages in arrears that are under restructuring with the mortgage providers. Switcher mortgages are exempt because the decision to move mortgage to a new provider does not increase the indebtedness of the borrower and does not restrict competition in the market. 

__Why do we need to implement?__ 
* Boom-busts in the housing market reflects: 
1. Higher leverage for borrowers than is typically the case for other investment activities. 

2. Speculative behaviour. 

3. Lags in supply response 

4. Slow pace of price discovery and illiquidity in the market. 

__What happens with high LTVs and LTIs__

1. Contributes to harmful boom bust cycles in housing markets
2. Weakening the resilience of banks and households to property market downturns. 

We find that LTV and LTI/DTI can be effective in reducing pro-cyclicality of house prices, credit, and leverage, and lowering default prob, and loss given default on mortgage credit. 

__Reducing Pro-cyclicality__ 
Increasing home value can allow for increased borrowing against the home value, which encourages pro-cyclicality. This can occur even in light of a LTV cap. However, LTV caps can limit the extent to pro-cyclicality. 

__What has literature found?__ 

LTV and LTI measures curb mortgage credit growth and reduce risk of house price bubbles emerging. 

See Lim et al. (2011). Finds that DTI is more effective in reducing housing credit. 

Macroprudential policies work better in boom phases than in busts. 

These measures reduce asset growth for banks during upswings and and stops declines in asset growth during contractionary periods. 

LTV caps effective in slowing mortgage credit growth and house price appreciation. They found that a 10% increase in LTV ratio is associated with a 13% increase in nominal house prices. 

Duca et al (2010) finds that a 8-11% increase in house prices from 10% increase in LTV for FTBs. 

Ahuja and Nabar (2011) find that reductions in LTV contributes to reductions in transaction volume and price growth with a lag and suggest that LTV tightening could affect property activitiy through the expectations channel rather than through the credit channel. 

Igan and Kang (2011) consider the expectations channel for these instruments and finds tighter limits on LTVs curb expectations and speculative incentives.

Regarding loss given default, restrictions provide a collateral buffer against house price declines. Lowers the loss in the event of default.
