__History and Business Overview__

* Founded in 1870
* Entered the news as Adelbert Delbruck's actions to xploit economic interdependencies between countries and circumvent the supremacy of British Banks that had dominated international foreign trade. 

-"Transact banking business of all kinds, in particular to promote and facilitate trade relations between Germany, other European countries, and overseas markets." 

- Was diversified in their cross-border commerical IB services within the same house. 

- First bank to adopt universal banking. Deepened economic links with the rest of the world.

- The bank handled international transactions for German companies and gobbled up tiny banks that couldn't survive the 1870 crisis. 

_Upon WW1_

* Germany inflation to the moon. 
* Deutsche bank lost many of its foreign assets. 
* Borrowers failed to pay back debts. 
* Bank had to sell its holdings.
* By 1929, a series of consolidation efforts resulted in deutsche bank merging with a rival Disconto-gesellschaft
* Before deutsche bank regained its title, inflation hit and the 1930 depression followed. 
* Nazis coincided with the economic and moral fall of deutsche bank. 
* German banks were broken down into smaller regions or nationalized. 
* Deutsche was pslit into 10 different banks and the name was outlawed. 
* 10 years later, remerged together.
* By 1958, Deutsche Bank issued first foregin currency bond in 44 years on the German capital markets. 
* Deutsche opened and acquired major banks in Italy, Spain, The UK, and US. Thirty years later, expanded into 12 countries. 
* Eventually acquired its way into 70 countries.

### The New Banking Environment and Basel III 

* Since 2009, fees from deal-making activities (M&A, E/D origination, etc) were down 25% over a two year period. Rev from Fixed Income ws down 25% since 2009 for the top 10 banks. 

* On top of this BASEL 3 would effectively increase capital requirements for these banks. 

* BASEL III changed T1 capital requirement from 4% of RWA to between 9.5% and 13.5% of RWA. Not only more capital, but more quality. 

* Regulation phased gradually starting 2013 and full compliance by 2019. 

* ROE expected to fall. 

* July 31, 2012: Core Tier 1 Ratio of 10.2% and total of 13.6% based on EUR373bn RWA. 

* Beginning 2013: Hade 7.2% T1 core ratio. Far from the 9.5% requirement. 

* Jain and Fitschen are now named co-CEOs. Jain previous head of IB, and Fitschen head ofo regional management.

* Both noted for their strengths in weathering German politics. 

* In late 2010, raised 9.8 BN EUR to acquire 50.2% of Postbank to enhance revenue mix and take a leading position iin the European retail banking business. By 2012, stake => 93.7%

* During the downturn in IB, consolidated with competitors or benefitted from their exits, and became the largest market share in U.S. fixed income trading, topping JP morgan chase. 

* Had internationla Presence
